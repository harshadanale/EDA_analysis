```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv("practice_3.csv")
ds
```

# employee count  per deoartment

```python
sns.countplot(data=ds,x='Department',palette='Set2')
plt.title('employee count  per deoartment')
```

finance department having more employee count follwed by marketing and hr department 

# Gender distribution

```python
sns.countplot(data=ds,x='Gender',palette='coolwarm')
plt.title("Gender distribution")
```

female having more count than male in the company

# employee location distribution

```python
sns.countplot(data=ds,x='Location',palette='Spectral')
plt.title('number of employee by location')
plt.xticks(rotation=45)
```


 seattle having more number of employee by location follwed by san francisco and new york 

# age distribution

```python
sns.histplot(ds['Age'],bins=10,kde=True)
plt.title("Age distribution")
```

 here 45 to 50 age having a more employee and than 25 to 30 age identify by the kde plot

# years of expriance distribution 

```python
sns.histplot(ds['Years_of_Experience'],bins=15,kde=True)
plt.title("years of experience")
```

 the max experience are in range of 2.5 year

# monthly salary distribution

```python
sns.boxplot(data=ds,y='Monthly_Salary',color='lightblue')
plt.title("monthly salary")
```

 monthly salary are minimum 40000 to  maximum 120000 and avg 75000

# avg salary by department

```python
sns.barplot(data=ds,x='Department',y='Monthly_Salary',estimator=np.mean,palette='viridis')
plt.title("avg salary by department")
```

 we can analysis the avg salary from the black lines of each block

# performance score by department 

```python
sns.boxplot(data=ds,x='Department',y='Performance_Score',palette='pastel')
plt.title(' performance score by department  ')
```

# salary distribution by gender

```python
sns.violinplot(data=ds,x='Gender',y='Monthly_Salary',palette='muted')
plt.title("salary distribution by gender")
```

female having minimum salary as compared to males

```python
sns.heatmap(ds[['Age','Years_of_Experience','Monthly_Salary','Performance_Score']].corr(), annot=True, cmap='coolwarm')
plt.title("correlationl matrix")
```

```python
sns.scatterplot(data=ds,x='Years_of_Experience',y='Monthly_Salary',hue='Department')
plt.title("salary vs. Experience by department ")
```

Finance department have increasing trend as YoE increases salary also increases followed by marketing and engineering department
